# Idea 1:

## Ask-Tell (BO-Lift)

```python
# Instantiate LLM model
bolift_at = bolift.AskTellFewShotTopk()
icl_examples = []
icl_values = []
# Tell
for _, row in mini_df.iterrows():
    icl_examples.append(row["compound id"])
    icl_values.append(row["solubility"])
    bolift_at.tell(row["compound id"], row["solubility"])
# Predict
bolift_pred = bolift_at.predict(molecule_name)
```

```python
# Find the MAE and Standard Deviation
if isinstance(bolift_pred, list):
    pred = bolift_pred[0].mean()
    bolift_mae = np.abs(molecule_sol - pred)
    bolift_std = bolift_pred.std()
else:
    pred = bolift_pred.mean()
    bolift_mae = np.abs(molecule_sol - pred)
    bolift_std = bolift_pred.std()

print(f"BO-Lift: MAE = {bolift_mae} | Standard Deviation = {bolift_std} | Prediction = {pred}.")
```

## Ask-Tell (QAFNet)

```python
# Instantiate QAFNet model
qaf_at = qafnet.QAFFewShotTopK()
# Tell
qaf_at.tell(data=mini_df)
# Predict
qaf_pred = qaf_at.predict(new_df.iloc[6][:-1])
```

```python
# Find the MAE and Standard Deviation
if isinstance(qaf_pred, list):
    pred = qaf_pred[0].mean()
    qaf_mae = np.abs(molecule_sol - pred)
    qaf_std = qaf_pred[0].std()
else:
    pred = qaf_pred.mean()
    qaf_mae = np.abs(molecule_sol - pred)
    qaf_std = qaf_pred.std
print(f"QAF-Net: MAE = {qaf_mae} | Standard Deviation = {qaf_std} | Prediction = {pred}.")
```

Given this ICL (5 examples), using model `text-babbage-001`, we can see that the QAFNet model, which incorporates more contextual information in the prompt, performs better than BO-Lift. The following experiments will look at different combinations of models and prompt inputs:

## Experiments

### Model (text-baggage-001)

```python
def run_bolift_experiment(num_examples: List[int], input_data: pd.DataFrame, pred_lim: int):
    # Instantiate LLM model
    bolift_at = bolift.AskTellFewShotTopk()
    # Tell
    _ = [bolift_at.tell(row["compound id"], row["solubility"]) for _, row in input_data.iloc[:num_examples].iterrows()] 
    # Predict
    bolift_preds = [bolift_at.predict(row["compound id"]) for _, row in input_data.iloc[num_examples:num_examples+pred_lim].iterrows()]
    # Find the MAE and Standard Deviation
    bolift_std = [pred[0].std() if isinstance(pred, list) else pred.std() for pred in bolift_preds]
    bolift_mse = np.mean([(row["solubility"] - pred[0].mean())**2 if isinstance(pred, list) else (row["solubility"] - pred.mean())**2 for pred in bolift_preds])
    return bolift_std, bolift_mse
```

```python
def run_qaf_experiment(num_examples: List[int], input_data: pd.DataFrame, pred_lim: int):
    # Instantiate LLM model
    qaf_at = qafnet.QAFFewShotTopK()
    # Tell
    _ = qaf_at.tell(input_data.head(num_examples))
    # Predict
    qaf_preds = [bolift_at.predict(row[:-1]) for _, row in input_data.iloc[num_examples:num_examples+pred_lim].iterrows()]
    # Find the MAE and Standard Deviation
    qaf_std = [pred[0].std() if isinstance(pred, list) else pred.std() for pred in qaf_preds]
    qaf_mse = np.mean([(row["solubility"] - pred[0].mean())**2 if isinstance(pred, list) else (row["solubility"] - pred.mean())**2 for pred in qaf_preds])
    return qaf_std, qaf_mse
```

```python
# Combine experiments
def run_all_experiments(num_examples: List[int], input_data: pd.DataFrame, num_iter: int, pred_lim: int):
    # Store data
    full_std_bolift = []
    full_mse_bolift = []
    full_std_qaf = []
    full_mse_qaf = []
    for j in tqdm(range(len(num_examples))):
        for _ in tqdm(range(num_iter)):
            data = input_data.sample(frac=1).reset_index(drop=True)
            bolift_std, bolift_mse = run_bolift_experiment(num_examples=num_examples[j], input_data=data, pred_lim=pred_lim)
            qaf_std, qaf_mse = run_qaf_experiment(num_examples=num_examples[j], input_data=data, pred_lim=pred_lim)
            full_std_bolift.append(bolift_std)
            full_mse_bolift.append(bolift_mse)
            full_std_qaf.append(qaf_std)
            full_mse_qaf.append(qaf_mse)
    return full_std_bolift, full_mse_bolift, full_std_qaf, full_mse_qaf
```

```python
# Run experiments
kwargs = {"num_examples": [5], "input_data": new_df, "num_iter": 10, "pred_lim": 5}
full_std_bolift, full_mse_bolift, full_std_qaf, full_mse_qaf = run_all_experiments(**kwargs)
```

```python
# Plot boxplot distribution of MSE
text_baggage_001_bp = [full_mse_bolift, full_mse_qaf]
# Creating a figure and a set of subplots
fig, ax = plt.subplots()
# Creating the boxplot
ax.boxplot(text_baggage_001_bp, showfliers=False)
# Setting the x-axis labels
ax.set_xticklabels(['BO-Lift', 'QAFNet'])
ax.set_xlabel("Algorithm")
ax.set_ylabel("MSE")
ax.set_title(f'Model = text_baggage_001 | Number of training examples = {kwargs["num_examples"][0]} | Number of prediction examples = {kwargs["pred_lim"]}')
# Displaying the plot
plt.show()
```

Prelminary results suggest that the contextual information (QAFNet( vastly reduces uncertainty in prediction however the error is greater compared to BO-Lift. This suggests that we may need to play around with:
- Model type.
- Number of training examples.
- Number of prediction examples.
- Number of contextual information pieces added.
